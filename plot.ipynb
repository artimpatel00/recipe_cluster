```python
from cluster_recipes.cluster_recipes import RecipeCluster
import plotly.express as px
```

```python
config_dict = {}
rc = RecipeCluster(config_dict)
```

```python
recipes_df = rc(use_saved_embeddings=True,
                sample_size=10000,
               cluster_max_dist=0.37,
               cluster_min_samples=15)
```

```python
print(recipes_df.shape)
recipes_df.head()
```

```python
fig = px.scatter(recipes_df, x="principal_component_one", y="principal_component_two",
                 color="cluster",
                 hover_data=['titles'])

fig.update_layout(
    autosize=False,
    width=1000,
    height=1000,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ))
fig.show()
```
